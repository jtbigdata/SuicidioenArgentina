```python
import pandas as pd
import numpy as np
```

```python
# Especifica la ruta de tu archivo CSV
ruta_archivo = "/home/julio/jupyter_files/ArgentinaSeguridad/suicidio/SAT-SS-BU_2017-2022.csv"

# Utiliza pd.read_csv() para importar el archivo CSV en un DataFrame
df = pd.read_csv(ruta_archivo)

# Ahora puedes trabajar con el DataFrame 'dataframe'
```

```python
df.info()
```

```python
# Utiliza dataframe.isnull().sum() para obtener la cantidad de valores nulos por columna
valores_nulos_por_columna = df.isnull().sum()

# Utiliza dataframe.shape[0] para obtener el número total de filas en el DataFrame
numero_total_filas = df.shape[0]

# Calcula el porcentaje de valores nulos por columna
porcentaje_valores_nulos_por_columna = (valores_nulos_por_columna / numero_total_filas) * 100

# Muestra el resultado
print('porcentaje_valores_nulos_por_columna')
print(porcentaje_valores_nulos_por_columna)
```

```python
#eliminamos -->tipo_lugar_otro,modalidad_otro,motivo_origen_registro_otro
```

```python
columnas_a_eliminar = ['tipo_lugar_otro', 'modalidad_otro', 'motivo_origen_registro_otro']
# Utiliza el método drop() para eliminar las columnas
df = df.drop(columnas_a_eliminar, axis=1)
```

```python
# Convertir la columna 'fecha_hecho' a formato de fecha con múltiples formatos
df['fecha_hecho'] = pd.to_datetime(df['fecha_hecho'], errors='coerce', format='%d-%m-%Y')

# Crear una nueva columna 'dias' extrayendo el día de la fecha
df['dias'] = df['fecha_hecho'].dt.day

# Imprimir la columna 'dias' para verificar los resultados
print(df['dias'].value_counts())
```

```python
df = df.drop(columns=['fecha_hecho'])
```

```python
print(df.hora_hecho.unique())
print('=================================')
print(df.hora_hecho.value_counts())
print('=================================')
print(df.hora_hecho.size)
```

```python
horas = df['hora_hecho'].str.split(':')
# Selecciona solo el primer elemento de cada lista que será la hora
horas = horas.str[0]
# Crea una nueva columna 'hora' con las horas extraídas
df['hora'] = horas
df = df.drop(columns=['hora_hecho'])
```

```python
#Todo aquellos cuyos ID se repite en mas de una ocasión, son igual salvo por las columnas suicida_sexo. Debido a que no son muchas filas, las eliminamos
```

```python
id_eliminar=[12826,10871,9068,25902,34058,13279,12976]    
# Utiliza la función de indexación booleana para crear una máscara de las filas que no contienen los IDs a eliminar
mascara = ~df['id_hecho'].isin(id_eliminar)

# Utiliza la máscara para filtrar el DataFrame y conservar solo las filas que no contienen los IDs a eliminar
dfi = df[mascara]
```

```python
# Utiliza la función de indexación booleana para crear una máscara de las filas que no contienen los IDs a eliminar
id_eliminar=['999-999-30','999-999-30','999-999-58','999-999-94','999-999-54','999-999-34','999-999-66','999-999-6', '999-999-38', '999-999-82','999-999-2']
mascara = ~dfi['localidad_id'].isin(id_eliminar)

# Utiliza la máscara para filtrar el DataFrame y conservar solo las filas que no contienen los IDs a eliminar
dfii = dfi[mascara]

# Intenta convertir la columna 'localidad_id' a tipo entero después de filtrar el DataFrame
dfii['localidad_id'] = pd.to_numeric(dfii['localidad_id'])
```
